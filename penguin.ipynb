```python
#importing dataset
import ssl
import certifi
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Create SSL context using certifi
ssl_context = ssl.create_default_context(cafile=certifi.where())

# Load the dataset manually from GitHub (Seaborn source)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(urllib.request.urlopen(url, context=ssl_context))

df.head()
```

```python
df.info()
 #null values summary
df.isnull().sum()
```

```python
 #null values summary
df.isnull().sum()
```

```python
#handling missing values

# it fills numerical column with either median mean or mode
df["body_mass_g"].fillna(df["body_mass_g"].median(),inplace=True)
df['sex'].fillna(df['sex'].mode()[0],inplace=True) #yaha p hum mode ke through bhar rahe h
df.dropna(inplace=True) #it drops all the rows with reamining null values
df.isnull().sum() #again null values ki summary 
```

```python
#encode the categories
#number of categories ko no of column banaega
#naya dataframe bana rahe h
df_encoded=pd.get_dummies(df,drop_first=True)
df_encoded.head()
```

```python
#FEATURE SCALING(data ko ek scale p lana h )

from sklearn.preprocessing import StandardScaler  # scaler import
scaler = StandardScaler()  # scaler object
# sirf numeric columns ko scale karna 
scaled_features = scaler.fit_transform(df_encoded.select_dtypes(include=np.number))
# scaled data ko wapas DataFrame me convert karke same column names dena
scaled_df = pd.DataFrame(scaled_features, columns=df_encoded.select_dtypes(include=np.number).columns)
scaled_df.head() 

# df_encoded me se sirf numerical columns (number type ke columns) select karte hain
# Aur unpe scaler fit (mean/std nikalta hai) + transform (scale karta hai) apply karte hain.
# Output ek numpy array hota hai jisme har numerical column standardized hota hai.
# Scaled features ko pandas DataFrame me convert karte hain
# Aur original numerical columns ke naam wapas assign karte hain for readability.

```
