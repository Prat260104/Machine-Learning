```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df.isnull().sum()
```

```python
df["age"].fillna(df["age"].median(), inplace=True)   
df["embark_town"].fillna(df["embark_town"].mode()[0], inplace=True) 
df["deck"].fillna(df["deck"].mode()[0], inplace=True) 
df["embarked"].fillna(df["embarked"].mode()[0], inplace=True) 
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df_encoded=pd.get_dummies(df,drop_first=True)
df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler  # scaler import
scaler = StandardScaler()  # scaler object
# sirf numeric columns ko scale karna 
scaled_features = scaler.fit_transform(df_encoded.select_dtypes(include=np.number))
# scaled data ko wapas DataFrame me convert karke same column names dena
scaled_df = pd.DataFrame(scaled_features, columns=df_encoded.select_dtypes(include=np.number).columns)
scaled_df.head() 
```
